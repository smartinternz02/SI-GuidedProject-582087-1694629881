#Pie chart

```python
print(df.columns)
```

```python
columns_of_interest = ['alcohol', 'total','speeding', 'not_distracted', 'no_previous',
       'ins_premium', 'ins_losses']
data = [df[column].sum() for column in columns_of_interest]
labels = columns_of_interest
plt.figure(figsize=(7,7))
plt.pie(data, labels=labels, autopct='%.3f%%')
plt.legend()
plt.show()
```

**Inference - From the plot we can infer that each sector of the pie chart represents the contribution of one of the selected columns to the total sum of all the columns. The size of each slice is proportional to the sum of values in that column relative to the total sum. The plt.legend() function adds a legend to the pie chart, providing labels for each slice**
